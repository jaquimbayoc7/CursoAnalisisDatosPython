### **2. Tipos de Impresiones** ⛪

```python
##formas de impresión en python
## 1. Forma
print('Hola quiero empanada!!!! :)')

## 2. Forma
print('Hola Quiero empanada:',2)

## 3. Forma - Interpolación String
a = 2
print(f'Hola Quiero empanada: {a}')

## 4. Forma - Interpolación String
print('Hola Quiero {} empanadas'.format(a))

## 5. Forma - Interpolación String
print('Hola Quiero %d empanadas' %a)

```

### **3. Condicionales** ⛪

```python
###Tipos de datos
a= 25 #int64
b= 3.1 #Float64
estado = False #bool
nombres = 'Julian Quimbayo' ##Object - Texto

###Condicionales - Simple
if a > b:
  print('a es mayor que b')

###Condicionales Compuestas
if a > b:
  print('a es mayor que b')
else:
  print('b es mayor que a')

###Condicionales Anidadas
if a > b:
  print('a es mayor que b')
elif a == b:
  print('a es igual que b')
else:
  print('b es menor que a')
```

### **4. Ciclos** ⛪

```python
###Ciclos

for i in range(11):
  print(f'Este es el valor de: {i}')

###Loop
# Imprimir los números del 0 al 9
i = 0
while i <= 10:
    print("Este es el valor de:", i)
    i += 1  # Incrementa i en 1
```
