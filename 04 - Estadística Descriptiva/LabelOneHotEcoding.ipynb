```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Dataset de ejemplo
data = {'Nivel': ['Primaria', 'Secundaria', 'Universidad', 'Primaria', 'Universidad','Preescolar']}
df = pd.DataFrame(data)

df['Nivel'].value_counts()

```

```python

# Aplicar Label Encoding
encoder = LabelEncoder()
df['Nivel_encoded'] = encoder.fit_transform(df['Nivel'])

df['Nivel_encoded'].value_counts()
```

```python
data = {'TipoDoc': ['CC', 'TI', 'CC', 'TI', 'RC','RC']}
df2 = pd.DataFrame(data)
```

```python
df2['TipoDoc'].value_counts()
```

```python
encoder = LabelEncoder()
df2['TipoDoc'] = encoder.fit_transform(df2['TipoDoc'])

df2['TipoDoc'].value_counts()
```

```python
import pandas as pd

# Dataset de ejemplo
data = {'País': ['Colombia', 'México', 'México', 'Colombia', 'México','México','México']}
df = pd.DataFrame(data)

# Aplicar One-Hot Encoding con pandas
df_encoded = pd.get_dummies(df, columns=['País'], dtype='int64')

df_encoded.drop(['País_Colombia'], axis=1, inplace= True)

print(df_encoded)
```
