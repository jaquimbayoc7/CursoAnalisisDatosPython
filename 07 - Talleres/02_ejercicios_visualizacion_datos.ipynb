### 1. Importar Librerías
Importa las librerías necesarias: `numpy`, `pandas`, `matplotlib.pyplot` (como `plt`) y `seaborn` (como `sns`).

```python
# Tu código aquí
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Crear el Dataset
Ejecuta la siguiente celda para crear los DataFrames que usaremos en los ejercicios.

```python
# Configuración de estilo
sns.set_theme(style='whitegrid')
np.random.seed(42)

# Dataset 1: Ventas mensuales
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas_2022 = [45, 52, 48, 65, 70, 85, 90, 88, 75, 80, 95, 110]
ventas_2023 = [50, 58, 55, 72, 78, 92, 98, 95, 82, 88, 105, 120]

df_ventas = pd.DataFrame({
    'Mes': meses,
    'Ventas_2022': ventas_2022,
    'Ventas_2023': ventas_2023
})

# Dataset 2: Datos de empleados
n_empleados = 150
df_empleados = pd.DataFrame({
    'Edad': np.random.normal(35, 8, n_empleados).astype(int).clip(22, 65),
    'Salario': np.random.normal(50000, 15000, n_empleados).clip(25000, 120000),
    'Años_Experiencia': np.random.normal(8, 4, n_empleados).clip(0, 30),
    'Departamento': np.random.choice(['Ventas', 'IT', 'Marketing', 'RRHH'], n_empleados),
    'Satisfaccion': np.random.randint(1, 11, n_empleados)
})

# Dataset 3: Productos
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares']
unidades_vendidas = [120, 450, 380, 210, 290]

df_productos = pd.DataFrame({
    'Producto': productos,
    'Unidades': unidades_vendidas
})

print('Datasets creados exitosamente')
print(f'\ndf_ventas: {df_ventas.shape}')
print(f'df_empleados: {df_empleados.shape}')
print(f'df_productos: {df_productos.shape}')
```

### 3. Gráfico de Líneas Simple
Crea un gráfico de líneas que muestre las ventas de 2022 (`Ventas_2022`) a lo largo de los meses. Usa `plt.plot()` y añade título y etiquetas a los ejes.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2022'])
plt.title('Ventas Mensuales de 2022')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.savefig('Grafico3.png', dpi=400)
plt.show()
```

```python
df_ventas['Ventas_2022'].describe().T
```

De acuerdo a la gráfica anterior de ventas 2022 por meses, se evidencia lo siguiente:
- El promedio de ventas del 2022 fue de 77.5.
- Los meses de mayor auge son (julio, octubre, noviembre y diciembre) con un promedio de 92.5.
- Los meses de menores ventas fueron: Enero, febrero y marzo esto posiblemente debido a comienzo del año y temas escolares y postvacaciones.

### 4. Gráfico de Líneas Múltiples
Crea un gráfico que muestre las ventas de 2022 y 2023 en la misma figura. Usa diferentes colores y añade una leyenda.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2022'], color='#9c27b0', label='2022')
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2023'], color='#c62828', label='2023')
plt.title('Ventas del 2022 vs 2023')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
df_ventas[['Ventas_2022','Ventas_2023']].describe().T
```

De acuerdo a la gráfica anterior evidenciamos lo siguiente:
- Las ventas del años 2023 siguen la tendencia del año 2022, con un aumento de 7.5 unidades de ventas.
- El aumento de ventas ha generado un aumento en la dispersión de los datos, con una poca posibilidad de predicción de ventas.

### 5. Gráfico de Barras Vertical
Crea un gráfico de barras vertical que muestre las unidades vendidas por producto usando `df_productos`. Añade título y etiquetas.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
plt.bar(df_productos['Producto'], df_productos['Unidades'])
plt.title('Unidades Vendidas por Producto')
plt.xlabel('Productos')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.show()
```

De acuerdo con la gráfica anterior:
- Los periféricos presentan mayor venta que los dispositivos principales.
- En julio, octubre, noviembre y diciembre estos periféricos se venden con mayor frecuencia.
- Se sugiere a gerencia, establecer dos estrategias:
1. Incrementar la fuerza de venta en los meses de baja rotación.
2. Mantener la tendencia en los 4 meses de mayor venta.

### 6. Gráfico de Barras Horizontal

### 7. Histograma
Crea un histograma de la distribución de edades de los empleados (`df_empleados['Edad']`). Usa 15 bins y añade título.

### 8. Gráfico de Dispersión (Scatter Plot)
Crea un gráfico de dispersión que muestre la relación entre `Años_Experiencia` (eje X) y `Salario` (eje Y) de `df_empleados`.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
plt.scatter(df_empleados['Años_Experiencia'], df_empleados['Salario'])
plt.title('Relación entre Años de Experiencia y Salario')
plt.xlabel('Años de Experiencia')
plt.ylabel('Salario')
plt.show()
```

De acuerdo con la gráfica anterior:

- Se evidencia a partir del año 12.5 en adelante un retiro de las personas y/o empleados de la empresa.
- El aumento de salario es inferior a los 80000 USD, solo 6 Personas.

### 9. Gráfico de Pastel (Pie Chart)
Crea un gráfico de pastel que muestre la proporción de unidades vendidas por producto. Usa `plt.pie()` y muestra los porcentajes.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
plt.pie(df_productos['Unidades'], labels=df_productos['Producto'], autopct='%1.1f%%', explode=[0, 0.1,0.1,0,0], startangle=240)
plt.title('Proporción de Unidades Vendidas por Producto')
plt.show()
```

Se evidencia en la gráfica de torta que las ventas estan distribuidas en un 77% entre auriculares, mouse y Teclado.

### 10. Box Plot (Diagrama de Caja)
Crea un box plot de los salarios por departamento usando `df_empleados`. Usa `sns.boxplot()`.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
sns.boxplot(x='Departamento', y='Salario', data=df_empleados)
plt.title('Salarios por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Salario')
```

Según el diagrama de caja, se evidencia un valor medio de salario para todas las áreas similar, no hay promoción y desde el punto de vista técnico hay ruido en la clasificación de IT.

### 11. Violin Plot

### 12. Histograma con Seaborn (Histplot)
Crea un histograma de la distribución de salarios usando `sns.histplot()`. Añade una curva KDE (Kernel Density Estimate).

### 14. Scatter Plot con Seaborn
Crea un scatter plot de `Edad` vs `Salario` usando `sns.scatterplot()`. Colorea los puntos según el `Departamento`.

### 15. Heatmap de Correlación
Crea un heatmap que muestre la matriz de correlación de las variables numéricas de `df_empleados`. Usa `sns.heatmap()` con `annot=True`.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
correlacion_matrix = df_empleados[['Edad', 'Salario','Años_Experiencia','Satisfaccion']].corr(method='spearman')
sns.heatmap(correlacion_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

### 16. Pair Plot

### 17. Subplots (Múltiples Gráficos)
Crea una figura con 2 subplots (1 fila, 2 columnas):
- Subplot 1: Histograma de edades
- Subplot 2: Histograma de salarios

Usa `plt.subplots()`.

```python
# Tu código aquí
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].hist(df_empleados['Edad'], bins=15)
axes[0].set_title('Distribución de Edades')
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Frecuencia')

axes[1].hist(df_empleados['Salario'], bins=15, color='orange')
axes[1].set_title('Distribución de Salarios')
axes[1].set_xlabel('Salario')
axes[1].set_ylabel('Frecuencia')

axes[2].hist(df_empleados['Edad'], bins=15)
axes[2].set_title('Distribución de Edades')
axes[2].set_xlabel('Edad')
axes[2].set_ylabel('Frecuencia')

axes[3].hist(df_empleados['Salario'], bins=15, color='orange')
axes[3].set_title('Distribución de Salarios')
axes[3].set_xlabel('Salario')
axes[3].set_ylabel('Frecuencia')


plt.show()
```
